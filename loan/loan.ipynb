plot 6 scatter charts at a time.

```python
def univariate_scatter(df, col1, col2, col3, col4, col5, col6,):
    fig, axes = plt.subplots(2, 3, figsize=(10, 8))
    y = df.index
    # Scatter plot 1
    axes[0, 0].scatter(df[col1], y)
    axes[0, 0].set_xlabel(col1)
    # Scatter plot 2
    axes[0, 1].scatter(df[col2], y)
    axes[0, 1].set_xlabel(col2)
    # Scatter plot 3
    axes[0, 2].scatter(df[col3], y)
    axes[0, 2].set_xlabel(col3)
    # Scatter plot 4
    axes[1, 0].scatter(df[col4], y)
    axes[1, 0].set_xlabel(col4)
    # Scatter plot 5
    axes[1, 1].scatter(df[col5], y)
    axes[1, 1].set_xlabel(col5)
    # Scatter plot 6
    axes[1, 2].scatter(df[col6], y)
    axes[1, 2].set_xlabel(col6)

    plt.tight_layout()
    plt.show()


univariate_scatter(df4, "loan_amnt","funded_amnt","funded_amnt_inv","term","int_rate","installment")
```

<h1>Multivariate Analysis

Descriptive analysis:

```python
import pandas as pd

def descriptive_analysis(df):
    # Calculate basic summary statistics for each column
    count = df.count()
    mean = df.mean()
    std_dev = df.std()
    minimum = df.min()
    percentile_25 = df.quantile(0.25)
    median = df.median()
    percentile_75 = df.quantile(0.75)
    maximum = df.max()
    mode = df.mode().iloc[0]
    
    # Combine the statistics into a single DataFrame
    result_df = pd.DataFrame({
        'Count': count,
        'Mean': mean,
        'Std Dev': std_dev,
        'Min': minimum,
        '25%': percentile_25,
        'Median': median,
        '75%': percentile_75,
        'Max': maximum,
        'Mode': mode
    })
    
    return result_df

descriptive_stats = descriptive_analysis(df5)
print(descriptive_stats)
```

```python
import seaborn as sns

def visualize_data(df):
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
    for col in numerical_cols:
        plt.figure(figsize=(4, 3))
        sns.histplot(data=df, x=col, kde=True)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
    
    sns.pairplot(df[numerical_cols])
    plt.title('Scatter Plot Matrix')
    plt.show()

visualize_data(descriptive_stats)
```
