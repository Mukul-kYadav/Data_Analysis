<h1>Data Cleaning

Deleting all the columns with NA values.

```python
missing_values = df.isna().sum()
missing_values
```

```python
drop_cols = missing_values[missing_values == 39717].index
df1 = df.drop(columns=drop_cols)
df1.head()
```

Deleting all the columns with 0

```python
columns_with_all_zeros = df1.columns[(df1 == 0).all()]
df2 = df1.drop(columns=columns_with_all_zeros)
```

```python
df2.head()
```

```python
df2.shape
```

Deleting columns with too much null values

```python
most_na = df2.isna().sum()
cols_with_na = most_na[most_na > 0]
cols_with_na
```

```python
df3 = df2.drop(columns=["mths_since_last_record", "next_pymnt_d"])
df3.shape
```

```python
df3.describe()
```

```python
df3.info()
```

```python
df3.head()
```

```python
import re
df3["term"] = df3["term"].apply(lambda x: int(re.findall(r'\d+', str(x))[0]) if pd.notnull(x) else x)
df3.head()
```

```python
df3["int_rate"] = df3["int_rate"].str.strip("%")
df3.head()
```
